```python
#A system of school that will input the result of the student
from time import sleep
data = []
student = {}
grade = 0
print("""=============================================
PROGRAMING SCHOOL SYSTEM - GRADES OF STUDENTS
=============================================""")
sleep(2)
print("Please, type the Name and the grades...\n")
sleep(0.7)

while True:
    student["Name"] = input("Name of the student: ") #Name
    for i in range(5):
        x = int(input(f"Grade {i+1} of the aluno: "))
        grade += x
    student["Grade"] = grade/5 #Grade
    if student["Grade"] >= 7:
        student["Approval"] = "Approved"
    else:
        student["Approval"] = "Disapproved" #Approval
    stop = str(input("Stop? [Yes/No]"))
    data.append(student)
    student = {}
    if stop == "Yes":
        break
    else:
        continue
print("="*45)
sleep(1)
print(f'|{"Name":<10}{"Media":<10}{"Approval"}\n|---------------------------------')
for i in range(len(data)):
    print(f'|{data[i]["Name"]:<10}{data[i]["Grade"]:<10}{data[i]["Approval"]}')
        
    
```

```python
#A program that will input things about a soccer player and show the performance
from time import sleep
Data = []
y = []
print("                 Player Register")
while True:
    print("+-"*30)
    Player = {"Name": input("Nome do jogador: "),
           "Matches": int(input("Quantas partidas no campeonato: "))}
    for i in range(Player["Matches"]):
        x = int(input(f"Quantos gols {Player['Name']} fez na {i+1}ª partida:"))
        y.append(x) 
    Player["Gols"] = y[:]
    Player["Total"] = sum(y)
    Data.append(Player)
    breaki = input("Wanna Stop? [S/N]").upper()
    if breaki == "S":
        break
print("+-"*30)
sleep(1)
print(f'|{"N.":<3}{"Nome":<10}{"Partidas":<10}{"Gols":<7} |')
print("-"*30)
for i in range(len(Data)):
    print(f"|{i:<3}{Data[i]['Name']:<10}{Data[i]['Matches']:<10}{Data[i]['Total']:<7} |")
print("+-"*30)
while True:
    about = int(input("Qual jogador deseja saber os gols(999 stops): "))
    if about == 999:
        break
    print(f"O jogador 0({Data[about]['Name']}) fez: ", end = " ")
    print(Data[about]["Gols"])
```

```python
#A system that will register some peoples and show some information about them
print("=-"*30)
y = ages = 0
x3 = [] 
x4 = []
data = []
while True:
    peoples ={"Name": input("Nome: "),
              "Sex": input("Sexo: [M/W] "),
              "age": int(input("Idade: "))}
    data.append(peoples)
    breaki = input("Wanna stop? [S/N]: ")
    if breaki == "S":
        break
#step 1
x1 = len(data)
#step 2
for i in range(len(data)):
    ages += data[i]["age"]
x2 = ages/len(data)
#step 3
for i in range(len(data)):
    if data[i]["Sex"] == "W":
        x3.append(data[i]["Name"])
#step 4
for i in range(len(data)):
    if data[i]["age"] > x2:
        x4.append(data[i]["Name"])

#Transformando W, M em Mulher e Homen
def gender_str(x):
    if x == "W":
        return("Mulher")
    else:
        return("Homen")

#Showing the information
print(f"""
Foram cadastradas {x1} pessoas...
A média de idade do grupo é {x2} anos...
As mulheres são: {x3}
As pessoas com idade acima da média são: {x4}""")
print("=-"*30)
print("As pessoas cadastradas foram...")
for i in range(len(data)):
    print(f"Nome: {data[i]['Name']};Sexo: {gender_str(data[i]['Sex'])};Idade: {data[i]['age']}")
print("=-"*30)    
```
