```python
#A program that you'll do a lista de compras que in the finish will show you some informations and the total
from time import sleep
h = 0
total = 0
morethan100 = 0
print("=-=-=-= ANALISE - LISTA DE COMPRAS =-=-=-=")
while True: 
    print("=-="*14)
    sleep(1)
    produto = str(input("Qual produto deseja cadastrar:    "))
    preço = int(input("Qual o preço de tal produto:   "))
    continuar = str(input("Quer continuar a lista? [S/N]  "))
    total =+ preço
    if preço >= 1000:
        morethan100 =+ 1        
    while h == 0:
        preçox = preço
        prudutox = produto
        h =+ 1
    if preço < preçox:
        preçox = preço
        produtox = produto
    if continuar == "N":
        break
print("\n","=-"*30, "\n")
print(f"O total da compra resultará em R{total}!!")
print(f"No total da lista, {morethan100} produtos custão mais de 100 reais!!")
#print(f"O produto mais barato da lista é o {produtox}, custando apenas {preçox} reais!!")
print("\n Volte sempre e tenha um ótimo dia :D")
```

```python
## A program that can solve a factorial number :)
numero = int(input("Digite um numero, na qual queira fatoriar:"))
fatorial = numero
contador = 1
while (numero-contador)>1:
    fatorial = fatorial*(numero-contador)
    contador += 1
print("{}! = {} ".format(numero,fatorial))
```

```python
#A try to do the sequence of fibonacci, o primeiro passo para contruir o triangulo de pascal :D
from time import sleep
a = 0
b = 1
H = 0
print("=-"*16)
print("    THE SEQUENCE OF FIBONACCI    ")
print("=-"*16)
print("\n")
T = int(input("Quantos termos quer avançar na sequencia: \n"))
while H <= T:
    if a == 0:
        print(a, end=", ")
        H += 1
    sleep(0.3)
    print(b, end=", ")
    H += 1
    sleep(0.3)
    c = a + b
    a = c
    print(a , end=", ")
    H += 1
    sleep(0.3)
    b = a + b
```
